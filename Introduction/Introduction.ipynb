## Why Chemical Engineering?

![Why chemical engineering?](why-icons-02.png)

Choosing a career in chemical engineering can be a rewarding and fulfilling decision for several reasons:

1. **Problem Solving**: Chemical engineers are problem solvers. They apply their knowledge of chemistry, physics, and mathematics to tackle real-world challenges and find innovative solutions. Whether it's optimizing industrial processes, designing new materials, or developing sustainable technologies, chemical engineers play a vital role in improving various industries.

2. **Diverse Career Opportunities**: Chemical engineers have a wide range of career opportunities across various industries. They can work in traditional sectors like oil and gas, pharmaceuticals, food processing, and petrochemicals. Additionally, they can contribute to emerging fields such as biotechnology, renewable energy, environmental engineering, and nanotechnology.

3. **Global Impact**: The work of chemical engineers has a global impact. They contribute to creating products and processes that enhance people's lives, improve health, and protect the environment. From developing life-saving medicines to designing cleaner energy solutions, chemical engineers make a positive difference in society.

4. **Innovation and Creativity**: Chemical engineering is a field that fosters innovation and creativity. As technology advances and new challenges arise, chemical engineers continuously seek creative ways to solve problems, invent new materials, and develop cutting-edge technologies.

5. **High Demand**: Chemical engineers are in high demand worldwide due to their versatile skill set and their ability to contribute to various industries. This demand often translates to competitive salaries and numerous job opportunities.

6. **Continuous Learning**: Chemical engineering is a dynamic field with constant advancements. As a chemical engineer, you'll have the opportunity to engage in lifelong learning, staying up-to-date with the latest research and technologies.

7. **Collaborative Work Environment**: Chemical engineers often work in collaborative teams, involving professionals from different disciplines. This teamwork fosters diverse perspectives and encourages a cooperative and stimulating work environment.

8. **Sustainability Focus**: In recent years, there has been an increasing emphasis on sustainability and environmental responsibility. Chemical engineers play a crucial role in developing green and sustainable technologies, helping to address global challenges such as climate change and resource depletion.

9. **Versatility**: A chemical engineering degree equips you with a versatile skill set that extends beyond traditional chemical engineering industries. The problem-solving, analytical, and technical skills gained can be applied in various roles, making chemical engineers adaptable to different career paths.

Ultimately, choosing chemical engineering as a career path can lead to a fulfilling and impactful professional journey, where you have the chance to make meaningful contributions to society and shape a better future.

## History of Chemical Engineering
Chemical engineering is a branch of engineering that deals with the application of physical and chemical processes to design, develop, and operate processes that transform raw materials into useful products. Here's a chronological overview of its history:

![History of Chemical Engineering](OldPicture1.jpg)

### 19th Century
* **Late 19th Century**: The roots of chemical engineering can be traced back to the late 19th century when industrial chemistry and chemical processes began to gain significance. Early chemical engineers focused on improving industrial processes, especially in the chemical and petrochemical industries.

### 20th Century
* **Early 20th Century**: The emergence of the industrial revolution saw a rapid expansion of chemical industries. As chemical processes became more complex, the demand for skilled engineers who could optimize and design these processes increased.

* **World War I and II**: The two world wars played a crucial role in the growth of chemical engineering. Chemical engineers contributed significantly to the production of explosives, fuels, medicines, and other materials needed for the war efforts.

* **1940s - 1950s**: With the development of new materials and synthetic polymers, chemical engineers played a key role in the plastics and synthetic fiber industries, which revolutionized various sectors like packaging, textiles, and electronics.

* **Late 20th Century**: Chemical engineering expanded beyond traditional industries to include areas such as environmental engineering, biotechnology, pharmaceuticals, and food processing. The focus shifted towards sustainable and environmentally friendly processes.

### 21st Century
* **21st Century**: Chemical engineering continues to evolve and adapt to the changing needs of society and technological advancements. With a greater emphasis on green and sustainable technologies, chemical engineers work on renewable energy, waste management, and climate change mitigation.

* **Advancements in Computing**: The advent of powerful computers and simulation software has enabled chemical engineers to model and optimize complex processes more effectively.

* **Interdisciplinary Collaboration**: Chemical engineers now collaborate with experts from various fields, such as biology, materials science, and nanotechnology, to address contemporary challenges.

Today, chemical engineering plays a critical role in improving our lives by producing essential products and developing innovative solutions to global challenges.

Please note that this is a concise overview, and the history of chemical engineering is vast and full of significant developments and contributions from numerous individuals and groups over the years.

## What differentiates chemical engineering from other engineering?

Chemical engineering is a branch of engineering that focuses on applying principles of chemistry, physics, biology, and mathematics to design, develop, and optimize processes that involve the transformation of raw materials into useful products or energy - Basically using processes to make stuff...


![Process flow diagramme](Process-Flow-Diagram-1024x538.jpg)

Several key factors differentiate chemical engineering from other engineering disciplines:

1. **Chemical Reactions and Processes**: Chemical engineering deals with chemical reactions and processes on a fundamental level. Understanding how reactions occur, their kinetics, and how to control them is a central aspect of chemical engineering.

2. **Chemistry and Material Science**: Chemical engineers work with a wide range of materials, including chemicals, polymers, metals, and more. They need a solid understanding of chemistry and material science to develop and design processes effectively.

3. **Mass and Energy Transfer**: Chemical engineers often work with systems involving the transfer of mass (e.g., fluids) and energy (e.g., heat transfer). This knowledge is essential in designing efficient and sustainable processes.

4. **Process Design and Optimization**: Chemical engineers are experts in process design, from laboratory-scale experimentation to large-scale industrial production. They aim to optimize processes for efficiency, safety, and cost-effectiveness.

5. **Safety and Environmental Considerations**: Chemical engineering involves handling potentially hazardous materials and processes. Chemical engineers must prioritize safety and ensure that their designs adhere to environmental regulations.

6. **Product Development**: Chemical engineers not only focus on process design but also play a role in product development. They work on creating new materials and products to meet specific needs.

7. **Interdisciplinary Nature**: Chemical engineering often requires collaboration with other engineering disciplines, such as mechanical, electrical, and environmental engineering, to achieve comprehensive solutions.

8. **Diverse Applications**: Chemical engineering has applications in various industries, including petrochemicals, pharmaceuticals, food and beverages, materials, energy, environmental protection, and more.

9. **Scale of Operations**: Chemical engineering involves designing processes that can operate at different scales, from laboratory benchtops to industrial-sized plants.

10. **Economic Considerations**: Chemical engineers must consider economic factors when designing processes. They strive to create processes that are not only technically feasible but also economically viable.

In summary, chemical engineering's focus on chemical reactions, material science, mass and energy transfer, process design, safety, and its interdisciplinary applications distinguish it from other engineering disciplines. The field's versatility allows chemical engineers to contribute to a wide range of industries and address various societal challenges.

## Why Chemical Reaction Engineering?

As mentioned before, Chemical Reaction Engineering is one of the key subdisciplines within Chemical Engineering. Chemical Reaction Engineering is a field of study that deals with the design and optimization of chemical reactors and processes. It emerged as a distinct discipline during the early 20th century and has since played a crucial role in various industries, including chemical, petrochemical, pharmaceutical, and environmental.

![Chemical Reactors](fogler_f01_01.jpg)

### Historical Background
The origins of Chemical Reaction Engineering can be traced back to the pioneering work of various scientists and engineers. Some key contributors to the field include:

1. **Paul Sabatier** (1854-1941) ![Paul Sabatier](Sabatier.jpeg)
A French chemist who made significant contributions to catalysis. He investigated the hydrogenation of organic compounds and discovered the catalytic properties of finely divided metals, earning him the Nobel Prize in Chemistry in 1912.



2. **Irving Langmuir** (1881-1957) ![Irving Langmuir](Langmuir.jpeg)
An American chemist and physicist who studied the adsorption of gases on solid surfaces. His work laid the foundation for understanding surface reactions and catalysis.

3. **Peter Waage** (1833-1900) ![Peter Waage](Waage.jpeg) and **Cato Guldberg** (1836-1902) ![Cato Guldberg](Guldberg.jpeg)
Norwegian chemists who formulated the law of mass action, which describes the rate of a chemical reaction as proportional to the product of the concentrations of the reactants.

3. **Octave Lievenspiel** (1926-2017) ![Octave Lievenspiel](Lievenspiel.jpeg)
American Chemical Engineer who wrote numerous textbooks on Chemical Reaction Engineering and developed the well-known "Lievenspiel-plots" used to optimise reactor sizes and configurations.

### Development as a Discipline
Chemical Reaction Engineering evolved as chemical engineering itself was established as an independent discipline in the early 20th century. The growing demand for chemical products and the need for efficient industrial processes drove the exploration of chemical reactions in controlled environments.

As chemical plants and industries expanded, the understanding of reaction kinetics, mass transfer, and reactor design became crucial for optimizing production processes, ensuring safety, and minimizing environmental impact. Chemical engineers began to apply principles from physics, chemistry, and mathematics to study and model reactions in various types of reactors.

### Modern Applications
Today, Chemical Reaction Engineering continues to be a vital field of study, with applications in numerous industries. From the production of pharmaceuticals to the synthesis of polymers, and from refining petroleum to designing environmentally friendly processes, the principles of Chemical Reaction Engineering play a fundamental role.

For a more in-depth exploration of the origin and development of Chemical Reaction Engineering, you can watch this informative YouTube video: [Origin of Chemical Reaction Engineering](https://www.youtube.com/watch?v=H-myZaCmtdg)
